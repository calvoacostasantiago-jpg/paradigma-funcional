### 1. Inicialización de Datos y Registro de Estudiante (Inmutable)

```python
# Importamos 'reduce' de functools para realizar agregaciones funcionales.
from functools import reduce

# La lista de estudiantes es inmutable en el sentido de que
# cada operación de 'modificación' devolverá una *nueva* lista de estudiantes.
# Inicialmente, está vacía.
students = []

# Función pura para registrar un estudiante.
# Recibe la lista actual de estudiantes, el nombre y la nota.
# Devuelve una *nueva* lista de estudiantes con el nuevo estudiante añadido.
# Esto asegura la inmutabilidad de la lista original.
def register_student(current_students, name, grade):
    # Creamos un nuevo diccionario para el estudiante.
    new_student = {'name': name, 'grade': grade}
    # Devolvemos una nueva lista que es la concatenación de la lista actual
    # y el nuevo estudiante. La lista original 'current_students' no se modifica.
    return current_students + [new_student]

# --- Demostración ---
# Registramos algunos estudiantes, asignando el resultado a 'students'
# para mantener la inmutabilidad y tener la lista actualizada.
students = register_student(students, 'Juan', 4.5)
students = register_student(students, 'Maria', 3.0)
students = register_student(students, 'Pedro', 2.8)
students = register_student(students, 'Ana', 5.0)

print("Estudiantes después del registro:")
# Mostramos la lista de estudiantes actualizada.
# Utilizaremos una función auxiliar para mostrar los estudiantes más adelante.
for student in students:
    print(f"- {student['name']}: {student['grade']}")
```

### 2. Mostrar Listado de Estudiantes (Función Pura)

```python
# Función pura para mostrar el listado de estudiantes.
# Toma la lista de estudiantes y, por cada uno, imprime sus datos.
# No modifica ningún estado externo y siempre produce el mismo resultado
# (la impresión) para la misma lista de entrada.
def list_students(students_list):
    print("\n--- Listado de Estudiantes ---")
    # 'map' no es estrictamente necesario aquí si solo vamos a imprimir,
    # pero lo usamos para demostrar su aplicación funcional.
    # La función lambda crea una cadena formateada para cada estudiante.
    # list() convierte el objeto map a una lista para forzar la evaluación si fuera necesario,
    # aunque en este caso, el bucle for es suficiente para iterar y mostrar.
    # Si no queremos retornar nada y solo queremos un efecto secundario (imprimir),
    # el bucle for es más idiomático. Sin embargo, para fines de demostración de map,
    # podríamos pensar en 'map' transformando los datos antes de un efecto.
    # En un contexto puramente funcional, una función de visualización
    # podría *retornar* una lista de cadenas a imprimir, en lugar de imprimir directamente.

    # Optamos por la impresión directa para la simplicidad de la salida al usuario.
    if not students_list:
        print("No hay estudiantes registrados.")
    else:
        # Usamos map para transformar cada estudiante en una cadena formateada.
        # Luego iteramos sobre estas cadenas para imprimirlas.
        formatted_students = map(lambda s: f"- Nombre: {s['name']}, Nota: {s['grade']}", students_list)
        for student_info in formatted_students:
            print(student_info)

# --- Demostración ---
list_students(students)
```

### 3. Calcular Promedio General (Usando `reduce`)

```python
# Función pura para calcular el promedio general de las notas.
# Utiliza 'reduce' para sumar todas las notas y luego divide por el número de estudiantes.
# No modifica la lista de estudiantes y siempre devuelve el mismo promedio
# para la misma lista de entrada.
def calculate_general_average(students_list):
    # Si no hay estudiantes, el promedio es 0 para evitar división por cero.
    if not students_list:
        return 0.0

    # Usamos map para extraer solo las notas de cada estudiante.
    grades = list(map(lambda s: s['grade'], students_list))

    # Usamos reduce para sumar todas las notas.
    # 'reduce' aplica la función lambda (acumulador + nota) a la lista de notas.
    # El '0' es el valor inicial del acumulador.
    total_grades = reduce(lambda acc, grade: acc + grade, grades, 0)

    # El promedio es la suma total de notas dividida por el número de estudiantes.
    average = total_grades / len(students_list)
    return average

# --- Demostración ---
general_average = calculate_general_average(students)
print(f"\nPromedio general de notas: {general_average:.2f}")
```

### 4. Mostrar Estudiante con Mayor Nota (Función Pura)

```python
# Función pura para encontrar al estudiante con la mayor nota.
# No modifica la lista de estudiantes y siempre devuelve el mismo estudiante
# (o None si la lista está vacía) para la misma entrada.
def get_highest_grade_student(students_list):
    # Si no hay estudiantes, no hay estudiante con la mayor nota.
    if not students_list:
        return None

    # La función 'max' puede tomar una clave ('key') para determinar cómo comparar elementos.
    # Aquí, la clave es una función lambda que devuelve la nota ('grade') de cada estudiante.
    # 'max' buscará el estudiante cuyo diccionario tenga el valor más alto en la clave 'grade'.
    highest_student = max(students_list, key=lambda s: s['grade'])
    return highest_student

# --- Demostración ---
highest_student = get_highest_grade_student(students)
if highest_student:
    print(f"\nEstudiante con la mayor nota: {highest_student['name']} ({highest_student['grade']})")
else:
    print("\nNo hay estudiantes para determinar la mayor nota.")
```

### 5. Mostrar Estudiantes Aprobados (Usando `filter`)

```python
# Función pura para obtener una lista de estudiantes aprobados.
# Utiliza 'filter' para seleccionar solo aquellos estudiantes cuya nota es mayor o igual a 3.0.
# Devuelve una *nueva* lista y no modifica la lista original.
def get_approved_students(students_list, min_grade=3.0):
    # 'filter' toma una función (aquí, una lambda) y un iterable.
    # Devuelve un iterador con los elementos para los que la función devuelve True.
    # La función lambda verifica si la nota del estudiante es mayor o igual a min_grade.
    approved = filter(lambda s: s['grade'] >= min_grade, students_list)
    # Convertimos el iterador 'filter' a una lista para poder trabajar con ella.
    return list(approved)

# --- Demostración ---
approved_students = get_approved_students(students)
print("\n--- Estudiantes Aprobados ---")
if not approved_students:
    print("No hay estudiantes aprobados.")
else:
    # Usamos la función list_students auxiliar o iteramos para mostrar.
    list_students(approved_students)
    # También podemos usar map para formatear y luego imprimir
    # formatted_approved = map(lambda s: f"- Nombre: {s['name']}, Nota: {s['grade']}", approved_students)
    # for student_info in formatted_approved:
    #     print(student_info)
```

### 6. Buscar Estudiante por Nombre (Usando `filter`)

```python
# Función pura para buscar estudiantes por nombre.
# Utiliza 'filter' para encontrar estudiantes cuyo nombre coincide (parcial o totalmente) con el buscado.
# Devuelve una *nueva* lista de estudiantes que coinciden y no modifica la lista original.
def search_student_by_name(students_list, search_name):
    # Convertimos el nombre de búsqueda a minúsculas para una comparación sin distinción de mayúsculas/minúsculas.
    search_name_lower = search_name.lower()

    # 'filter' selecciona estudiantes donde el nombre del estudiante (también en minúsculas)
    # contiene la cadena de búsqueda.
    found_students = filter(lambda s: search_name_lower in s['name'].lower(), students_list)
    # Convertimos el iterador 'filter' a una lista.
    return list(found_students)

# --- Demostración ---
print("\n--- Búsqueda de Estudiante por Nombre (Buscando 'an') ---")
found_by_name = search_student_by_name(students, 'an')
if not found_by_name:
    print("No se encontraron estudiantes con ese nombre.")
else:
    list_students(found_by_name)

print("\n--- Búsqueda de Estudiante por Nombre (Buscando 'juan') ---")
found_by_name_exact = search_student_by_name(students, 'juan')
if not found_by_name_exact:
    print("No se encontraron estudiantes con ese nombre.")
else:
    list_students(found_by_name_exact)
```
